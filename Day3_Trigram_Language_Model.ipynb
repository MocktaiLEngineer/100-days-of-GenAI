```python
names = open('names.txt','r').read().splitlines()
names
```

```python
# Build the vocab
TOKEN = '.'
vocab = [TOKEN] + sorted(set(''.join(names)))
vocab
```

```python
char_to_int = {char:i for i,char in enumerate(vocab)}
int_to_char = {v:k for k,v in char_to_int.items()}

char_to_int, int_to_char
```

```python
# Layout the structure of the trigram table
n = len(vocab)
trigram_table = torch.zeros(n,n,n)
trigram_table
```

```python
trigram_table.shape
```

```python
for name in names:
    word = [TOKEN] * 2 + list(name) + [TOKEN]
    for ch1,ch2,ch3 in zip(word,word[1:],word[2:]):
        ix1 = char_to_int[ch1]
        ix2 = char_to_int[ch2]
        ix3 = char_to_int[ch3]
        trigram_table[ix1,ix2,ix3] += 1
```

```python
char_to_int
```

```python
# Let's check if we've build the trigram table correctly
def check_for_correctness(char1,char2,char3):
    ix1 = char_to_int[char1]
    ix2 = char_to_int[char2]
    ix3 = char_to_int[char3]
    return trigram_table[ix1,ix2,ix3].item()

ans1 = check_for_correctness(".","a","b") #How many times was 'a' the starting character in the names dataset 

ans2 = sum([1 for name in names if name[:2] == "ab"])

ans1,ans2
```

```python
# Converting the trigram table into probabilities
P = (trigram_table + 1).float() #Applying smoothing
P /= P.sum(2, keepdims = True)
```

```python
# Generate new names using the bigram table
generator = torch.Generator().manual_seed(2147483647)

for i in range(10):
    name = []
    
    input_seed = '.'
    
    ix1 = char_to_int[input_seed]
    ix2 = char_to_int[input_seed]

    while True:
        probs = P[ix1,ix2]
        
        ix1 = ix2

        ix2 = torch.multinomial(probs, num_samples = 1, replacement = True, generator = generator).item()
        name.append(int_to_char[ix2])

        if ix2 == 0:
            break
    print(''.join(name))
```

```python
# How good is our model?

likelihood = 1
log_likelihood = 0

n = 0
for name in names:
    word = [TOKEN] + list(name) + [TOKEN]
    for ch1,ch2,ch3 in zip(word,word[1:],word[2:]):
        ix1 = char_to_int[ch1]
        ix2 = char_to_int[ch2]
        ix3 = char_to_int[ch3]
        prob = P[ix1,ix2,ix3]
        likelihood *= prob
        log_likelihood += torch.log(prob)
        n += 1

print(f"{likelihood=}")
print(f"{log_likelihood=}")

negative_log_likelihood = -log_likelihood
normalized_negative_log_likelihood = negative_log_likelihood / n

print(f"{negative_log_likelihood=}")
print(f"{normalized_negative_log_likelihood=}")
```

```python
# 2.09, that's like the score we would like to achieve with a NN based approach and we probably cannot go lower
```
